```python
#Random Funtion

import random

pickup_node = []
drop_node = []
for i in range(1000):
    c = random.randint(1,143)
    pickup_node.append(c)
    m = 1 ;
    while(True):
        m = random.randint(1,143)
        if m != c:
            break
    drop_node.append(m)
```

```python
path = []
for i in range(1000):
    x = pickup_node[i]
    y = drop_node[i]
    if x != 60:
        if y != 60:
            path.append(nx.shortest_path(G_removed,str(x),str(y)))
```

```python
counter = 0
for x in path:
    print(x)
    print("\n")
    counter = counter+1
print(counter)
```

```python
Dict = {}
nodes = list(G_removed.nodes())
for x in nodes:
    Dict[x] = 0
for pather in path:
    for i in pather:
        c = Dict[i]
        Dict[i] = c+ 1
print(Dict)
```

```python
import numpy as np
N = len(xs)
sizevalues = list(Dict.values())
color = np.random.randint(1, 5, size=N)
fig ,ax = plt.subplots(figsize=(10,10))
scatter = ax.scatter(xs,ys,s=sizevalues,c=color)
legend1 = ax.legend(*scatter.legend_elements(),loc="lower left", title="Classes")
ax.add_artist(legend1)
handles, labels = scatter.legend_elements(prop="sizes", alpha=0.6)
legend2 = ax.legend(handles, labels, loc="upper right", title="Sizes")
```
